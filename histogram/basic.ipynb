# **geom_histogram**

#### input ที่ต้องการ = ตัวแปร numeric 1 ตัวแปร

```r
# library
library(ggplot2)
 
# dataset:
data=data.frame(value=rnorm(100))

# basic histogram
ggplot(data, aes(x=value)) + 
  geom_histogram()
  options(repr.plot.width  = 7, repr.plot.height = 4, repr.plot.res = 200)
```

```r
# Libraries
library(tidyverse)
library(hrbrthemes)

# Load dataset from github
data <- read.table("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv", header=TRUE)

# plot
p <- data %>%
  filter( price<300 ) %>%
  ggplot( aes(x=price)) +
    geom_histogram( binwidth=3, fill="#69b3a2", color="#e9ecef", alpha=0.9) +
    ggtitle("Bin size = 3") +
    theme_ipsum() +
    theme(
      plot.title = element_text(size=15)
    )
p

p4 = data %>%
  filter( price<300 ) %>%
  ggplot( aes(x=price)) +
    geom_histogram( binwidth=50, fill="#69b3a2", color="#e9ecef", alpha=0.9) +
    ggtitle("Bin size = 3") +
    theme_ipsum() +
    theme(
      plot.title = element_text(size=15)
    )
p4
```

กำหนดขนาด bar ด้วยฟังก์ชั่น binwidth
