# **Mirror density chart with ggplot2**

#### ใช้เปรียบเทียบข้อมูลแบบ numeric 2 ตัวแปร

### **Density with geom_density**

```r
# Libraries
library(ggplot2)
library(hrbrthemes)

# Dummy data
data <- data.frame(
  var1 = rnorm(1000),
  var2 = rnorm(1000, mean=2)
)

# Chart
p <- ggplot(data, aes(x=x) ) +
  # Top
  geom_density( aes(x = var1, y = ..density..), fill="#69b3a2" ) +
  geom_label( aes(x=4.5, y=0.25, label="variable1"), color="#69b3a2") +
  # Bottom
  geom_density( aes(x = var2, y = -..density..), fill= "#404080") +
  geom_label( aes(x=4.5, y=-0.25, label="variable2"), color="#404080") +
  theme_ipsum() +
  xlab("value of x")
  options(repr.plot.width  = 9, repr.plot.height = 4, repr.plot.res = 300)
p
```

### **Histogram with geom_histogram**

#### วิธีทำเหมือน geom_density( ) แต่ทำเป็น geom_histogram( ) แทน

```r
# Libraries
library(ggplot2)
library(hrbrthemes)

# Dummy data
data <- data.frame(
  var1 = rnorm(1000),
  var2 = rnorm(1000, mean=2)
)

# Chart
p <- ggplot(data, aes(x=x) ) +
  # Top
  geom_histogram( aes(x = var1, y = ..density..), fill="#69b3a2") +
  geom_label( aes(x=4.5, y=0.25, label="variable1"), color="#69b3a2") +
  # Bottom
  geom_histogram( aes(x = var2, y = -..density..), fill= "#404080") +
  geom_label( aes(x=4.5, y=-0.25, label="variable2"), color="#404080") +
  theme_ipsum() +
  xlab("value of x")
  options(repr.plot.width  = 9, repr.plot.height = 4, repr.plot.res = 200)
p
```
