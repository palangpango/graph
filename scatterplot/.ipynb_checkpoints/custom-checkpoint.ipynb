# **Custom scatterplot**

#### เราสามารถเปลี่ยนรูปแบบของ plot ได้ดังต่อไปนี้ โดยใส่เข้าไปในฟังก์ชั่น **geom_point**

> #### **color** = ใส่สีเส้นขอบของจุด

> #### **Stroke** = ความกว้างของจุด

> #### **Fill** = ใส่สีภายในจุด

> #### **Shape** = รูปร่างของจุด

> #### **Alpha** = ความโปร่งแสงของจุด มีค่าตั้งแต่ 0 - 1 (0 โปร่งแสงมากที่สุด)

> #### **Size** = ขนาดของจุด

```r
# library
library(ggplot2)
 
# Iris dataset is natively provided by R
#head(iris)
 
# use options!
ggplot(iris, aes(x=Sepal.Length, y=Sepal.Width)) + 
    geom_point(
        color="black",
        fill="#FF6600",
        shape=21,
        alpha=0.5,
        size=8,
        stroke = 2
        )
```

### สามารถใช้ **theme_ipsum** ได้

```r
# library
library(ggplot2)
library(hrbrthemes)

# Iris dataset is natively provided by R
#head(iris)
 
# use options!
ggplot(iris, aes(x=Sepal.Length, y=Sepal.Width)) + 
    geom_point(
        color="black",
        fill="#69b3a2",
        shape=22,
        alpha=0.5,
        size=6,
        stroke = 1
        ) +
    theme_ipsum()
```
