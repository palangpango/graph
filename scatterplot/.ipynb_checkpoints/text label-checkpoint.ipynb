# **Add text and labels**

### **การเพิ่มตัวหนังสือบนกราฟ โดย geom_text()**

> #### **label** = ข้อความที่ต้องการให้แสดง

> #### **nudge_x และ nudge_y** = เลื่อนข้อความตามแนวแกน x และแกน y

> #### **Check_overlap** = ตรวจสอบไม่ให้ข้อความซ้อนทับกัน 

```r
# library
library(ggplot2)
 
# Keep 30 first rows in the mtcars natively available dataset
data=head(mtcars, 30)
 
# 1/ add text with geom_text, use nudge to nudge the text
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + # Show dots
  geom_text(
    label=rownames(data), 
    nudge_x = 0.25, nudge_y = 0.25, 
    check_overlap = T
  )
```

### **การเพิ่ม label บนกราฟ โดย geom_label**

```r
install.packages("ggrepel")
# library
library(ggplot2)
 
# Keep 30 first rows in the mtcars natively available dataset
data=head(mtcars, 30)
 
# 1/ add text with geom_text, use nudge to nudge the text
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + # Show dots
  geom_label(
    label=rownames(data), 
    nudge_x = 0.25, nudge_y = 0.25, 
    check_overlap = TRUE
  )
```

### **การเพิ่ม Text label เพียง 1 อัน**

> #### เราสามารถกำหนดรูปแบบของ text label เองได้ด้วยฟังก์ชั่น เช่น label.padding, label.size, color and fill 

```r
# library
library(ggplot2)
 
# Keep 30 first rows in the mtcars natively available dataset
data=head(mtcars, 30)
 
# Add one annotation
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + # Show dots
  geom_label(
    label="Look at this!", 
    x=4.1,
    y=20,
    label.padding = unit(0.55, "lines"), # Rectangle size around label
    label.size = 0.50,
    color = "black",
    fill="#FFB6C1"
  )
```

### **การเพิ่ม Text label ตรงข้อมูลที่มีเงื่อนไขที่เราต้องการ**

> #### โดยการ filter ข้อมูลที่เราต้องการ

```r
# library
library(ggplot2)
library(dplyr)
library(tibble)

# Keep 30 first rows in the mtcars natively available dataset
data=head(mtcars, 30)

# Change data rownames as a real column called 'carName'
data <- data %>%
  rownames_to_column(var="carName")
  
# Plot
ggplot(data, aes(x=wt, y=mpg)) +
  geom_point() + 
  geom_label( 
    data=data %>% filter(mpg>20 & wt>3), # Filter data first
    aes(label=carName)
  )
```
