# **Basic scatterplot**

> #### แสดงความสัมพันธ์ของ 2 ตัวแปร

#### **ขั้นตอนในการทำ scatterplot**

> #### 1.ระบุ dataframe

> #### 2.ระบุตัวแปรที่จะแสดงบนแกน x และ y

> #### 3. ใส่ geom_point () 

```r
# library
library(ggplot2)
 
# The iris dataset is provided natively by R
#head(iris)
 
# basic scatterplot
ggplot(iris, aes(x=Sepal.Length, y=Sepal.Width)) + 
    geom_point()
```
