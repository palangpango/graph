## **Adjacency matrix**

- #### คือเมตริกซ์ที่ข้อมูลใน row และ column เหมือนกัน และมีขนาดเท่ากัน

```r

#library
library(igraph)

# Create data
set.seed(10)
data <- matrix(sample(0:2, 25, replace=TRUE), nrow=5)
colnames(data) = rownames(data) = LETTERS[1:5]
 
# build the graph object
network <- graph_from_adjacency_matrix(data)
 
# plot it
plot(network)
```

## **Incidence matrix**

- #### คือเมตริกซ์ที่ข้อมูลใน row และ column ไม่เหมือนกันและมีขนาดไม่เท่ากัน

```r
# lib
library(igraph)

# data
set.seed(1)
data <- matrix(sample(0:2, 15, replace=TRUE), nrow=3)
colnames(data) <- letters[1:5]
rownames(data) <- LETTERS[1:3]
 
# create the network object
network <- graph_from_incidence_matrix(data)
 
# plot it
plot(network)
```

### **Edge list**

- #### มีข้อมูล 2 column แต่ละแถวจะบอกการเชื่อมต่อข้อมูลจากต้นทางสู่ปลายทาง

```r
# create data:
links <- data.frame(
    source=c("A","A", "A", "A", "A","F", "B"),
    target=c("B","B", "C", "D", "F","A","E")
    )

# create the network object
network <- graph_from_data_frame(d=links, directed=F) 

# plot it
plot(network)
```

### **Literal List of connections**

- #### คือการสั่งให้โปรแกรมเชื่อมโยงตัวอักษรที่เราต้องการได้

```r
# create data:
network <- graph_from_literal( A-B-C-D, E-A-E-A, D-C-A, D-A-D-C )

# plot it
plot(network)
```
