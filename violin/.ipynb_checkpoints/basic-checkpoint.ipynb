# **BASIC VIOLIN PLOT**

#### ด้วย ggplot ใช้ฟังก์ชั่น geom_violin()

```r
# Library
library(ggplot2)

# create a dataset
data <- data.frame(
  name=c( rep("A",500), rep("B",500), rep("B",500), rep("C",20), rep('D', 100)  ),
  value=c( rnorm(500, 10, 5), rnorm(500, 13, 1), rnorm(500, 18, 1), rnorm(20, 25, 4), rnorm(100, 12, 1) )
)

# Most basic violin chart
ggplot(data, aes(x=name, y=value, fill=name)) + # fill=name allow to automatically dedicate a color for each group
  geom_violin()
  options(repr.plot.width  = 5, repr.plot.height = 4, repr.plot.res = 200)
```

## **note**

#### **ggplot2**  ต้องการข้อมูลแบบ **long format**(แต่ละแถวมีเพียง 1 ข้อมูลเท่านั้น)

#### input ที่จำเป็นต้องใส่ คือ 1. ตัวแปรในแกน x เป็น factor 2. ตัวแปรในแกน y เป็น numeric
               

```r
# library
library(ggplot2)
library(dplyr)

# create data
data <- data.frame(
  name=c( rep("A",500), rep("B",500), rep("B",500), rep("C",20), rep('D', 100)  ),
  value=c( rnorm(500, 10, 5), rnorm(500, 13, 1), rnorm(500, 18, 1), rnorm(20, 25, 4), rnorm(100, 12, 1) ) %>% round(2)
)
# Basic violin
ggplot(data, aes(x=name, y=value, fill=name)) + 
  geom_violin()
options(repr.plot.width  = 9, repr.plot.height = 4, repr.plot.res = 300)
```

###  **สำหรับข้อมูลแบบ wide format ให้ใช้ฟังก์ชั่น gather() ของ tidyverse library**

```r
# Let's use the iris dataset as an example:
data_wide <- iris[ , 1:4]
```

```r
library(tidyr)
library(ggplot2)
library(dplyr)
data_wide %>% 
  gather(key="MesureType", value="Val") %>%
  ggplot( aes(x=MesureType, y=Val, fill=MesureType)) +
    geom_violin()
```
