## **Grouped barchart**

- #### ใส่ **position = "dodge"** เพื่อให้แยกกราฟออกเป็นแท่งตามกลุ่มข้อมูล

- #### ฟังก์ชั่น **aes()** ภายในให้ใส่เงื่อนไขที่ต้องการให้แยกข้อมูล ,แกน x,แกน y


```r
##### # library
library(ggplot2)
 
# create a dataset
specie <- c(rep("sorgho" , 3) , rep("poacee" , 3) , rep("banana" , 3) , rep("triticum" , 3) )
condition <- rep(c("normal" , "stress" , "Nitrogen") , 4)
value <- abs(rnorm(12 , 0 , 15))
data <- data.frame(specie,condition,value)
 
# Grouped
ggplot(data, aes(fill=condition, y=value, x=specie)) + 
    geom_bar(position="dodge",stat="identity")
    options(repr.plot.width  = 25, repr.plot.height = 7, repr.plot.res = 120)
```

## **Stacked barchart**

- #### การแบ่งข้อมูลเป็นชั้นๆ

- #### ใส่ **position = "stack"** เพื่อให้แยกกราฟออกเป็นชั้นตามกลุ่มข้อมูล

- #### ฟังก์ชั่น **aes()** ภายในให้ใส่เงื่อนไขที่ต้องการให้แยกข้อมูล ,แกน x,แกน y


```r
 # library
library(ggplot2)
 
# create a dataset
specie <- c(rep("sorgho" , 3) , rep("poacee" , 3) , rep("banana" , 3) , rep("triticum" , 3) )
condition <- rep(c("normal" , "stress" , "Nitrogen") , 4)
value <- abs(rnorm(12 , 0 , 15))
data <- data.frame(specie,condition,value)
 
# Stacked
ggplot(data, aes(fill=condition, y=value, x=specie)) + 
    geom_bar(position="stack", stat="identity")
```

## **Percent stacked barchart**

- #### การแสดงข้อมูลแบบ stack แต่แสดงในรูปร้อยละ

- #### ใส่ **position = "fill"** เพื่อให้แสดงร้อยละของข้อมูลในแต่ละกลุ่ม

```r
 # library
library(ggplot2)
 
# create a dataset
specie <- c(rep("sorgho" , 3) , rep("poacee" , 3) , rep("banana" , 3) , rep("triticum" , 3) )
condition <- rep(c("normal" , "stress" , "Nitrogen") , 4)
value <- abs(rnorm(12 , 0 , 15))
data <- data.frame(specie,condition,value)
 
# Stacked + percent
ggplot(data, aes(fill=condition, y=value, x=specie)) + 
    geom_bar(position="fill", stat="identity")
```

## **การกำหนดลักษณะของ Grouped barchart**

- #### ใส่ชื่อกราฟ ธีม เปลี่ยนสี เปลี่ยนชื่อแกน

```r
# library
library(ggplot2)
library(viridis)
library(hrbrthemes)

# create a dataset
specie <- c(rep("sorgho" , 3) , rep("poacee" , 3) , rep("banana" , 3) , rep("triticum" , 3) )
condition <- rep(c("normal" , "stress" , "Nitrogen") , 4)
value <- abs(rnorm(12 , 0 , 15))
data <- data.frame(specie,condition,value)
 
# Small multiple
ggplot(data, aes(fill=condition, y=value, x=specie)) + 
    geom_bar(position="stack", stat="identity") +
    scale_fill_viridis(discrete = T) +
    ggtitle("Studying 4 species..") +
    theme_ipsum() +
    xlab("")
```

## **Small multiple**

- #### การแยกกราฟออกมาเป็นกราฟเล็กๆตามกลุ่มข้อมูล 

- #### โดยใช้ฟังก์ชั่น **facet_wrap**

```r
# library
library(ggplot2)
library(viridis)
library(hrbrthemes)

# create a dataset
specie <- c(rep("sorgho" , 3) , rep("poacee" , 3) , rep("banana" , 3) , rep("triticum" , 3) )
condition <- rep(c("normal" , "stress" , "Nitrogen") , 4)
value <- abs(rnorm(12 , 0 , 15))
data <- data.frame(specie,condition,value)
 
# Graph
ggplot(data, aes(fill=condition, y=value, x=condition)) + 
    geom_bar(position="dodge", stat="identity") +
    scale_fill_viridis(discrete = T, option = "E") +
    ggtitle("Studying 4 species..") +
    facet_wrap(~specie) +
    theme_ipsum() +
    theme(legend.position="none") +
    xlab("")
```
