# **Basic Heatmap โดย Geom_tile**

#### Input ต้องเป็นแบบ Long format ต้องการ 3 ตัวแปร/observation

> #### 1) x = ตัวแปรบนแกน x

> #### 2) y = ตัวแปรบนแกน y

> #### 3) Fill = ค่าตัวเลขที่จะแปลเป็นสี

```r
# Library
library(ggplot2)

# Dummy data
x <- LETTERS[1:20]
y <- paste0("var", seq(1,20))
data <- expand.grid(X=x, Y=y)
data$Z <- runif(400, 0, 5)
 
# Heatmap 
ggplot(data, aes(X, Y, fill= Z)) + 
  geom_tile()
```
